```python
import numpy as np
arr=np.random.random((10,1))
print(arr)
m=np.mean(arr)
print(m)
```

```python
import numpy as np
arr=np.random.random((20,1))
print(arr)
sd=np.std(arr)
print("Standard Deviation:", sd)
print("Variance:",sd*sd)
```

```python
import numpy as np
A=np.random.random((10,20))
B=np.random.random((20,25))
C=np.matmul(A,B)
print(C)
print("Sum of elements of C:",np.sum(C))
```
